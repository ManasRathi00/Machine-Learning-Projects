```python
df_loan['risky_loan_applicant']=df_loan['loan_amnt']<=df_loan['funded_amnt']
```

```python
x=df_loan['risky_loan_applicant']
def risky_loan(x):
    if x==True:
        return 0
    else:
        return 1
```

```python
df_loan['risky_loan_applicant']=df_loan['risky_loan_applicant'].apply(risky_loan)
```

```python
x=df_loan['emp_length']
def emp_exp(x):
    if x<=1:
        return 'Fresher'
    elif x>1 and x<3:
        return 'Junior'
    elif x>3 and x<7:
        return 'Senior'
    else:
        return 'Expert'
```

```python
df_loan['emp_length']=df_loan['emp_length'].apply(emp_exp)
```
