Highly unbalanced dataset.


0 ---> Normal


1 ---> Fraudulent

```python
legit =df[df['Class'] == 0]
fraud =df[df['Class'] == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both
df.groupby('Class').mean()
```

Using Undersampling because the data is unbalanced.

Building a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions. 

Number of fraudulent transactions ---> 492


```python
legit_sample = legit.sample(n = 492)
```

```python
#Concatenation two dataframces
sample = pd.concat([legit_sample,fraud],axis = 0)
```

```python
sample.head()
```

```python
sample.tail()
```

```python
sample.shape
```

```python
sample.Class.value_counts()
```

```python
sample.groupby('Class').mean()
```

Splitting the data into features and targets

```python
X = sample.drop(columns = 'Class', axis = 1)
Y = sample.Class
```

```python
X.head()
```

```python
Y
```

Split the Data into Training Data and Testing Data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
#training the logisitc regression model with training data
print(model.fit(X_train, Y_train))
```

Model Evaluation 

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data',training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(x_test_prediction,Y_test)
```

```python
print('Accuracy on test data',test_data_accuracy)
```
